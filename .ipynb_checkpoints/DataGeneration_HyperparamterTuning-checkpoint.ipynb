```python
#generate class data

from  sklearn.datasets import make_classification as makeclass

classdata1 = makeclass(n_samples=14000, n_features=12, n_informative=10, n_redundant=0, 
          n_repeated=0, n_classes=6, n_clusters_per_class=3, weights=None, flip_y=0.02, 
          class_sep=0.5, hypercube=False, shift=2, scale=1.0, shuffle=True, random_state=178902)


# classdata1
X= np.asarray(classdata1[0])
y= np.asarray(classdata1[1])

X.shape

print(Counter(y))
print(stats.describe(X[:,2]))
```

```python
#subset and split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.2, random_state = 0)

#scale data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
# Permutation function 
def hyper_parameters(act, nod, batch_size,lay):   
    lists = [act, nod, batch_size,lay]
    permutations = list(itertools.product(*lists))
    return permutations

#lists
activation = ['sigmoid', 'relu', 'LeakyReLU']
nodes = [10, 20, 50]
batch_size = [8, 16, 32]
layers = [1,2]

#execution of function
permutations = hyper_parameters(activation,nodes,batch_size,layers)

#length of all combinations
permutations
print(len(permutations))
```

```python
def create_model_phase_one(X_train, X_test, y_train, y_test,a,b,c,d):
    
    #contruct model
    
    model = models.Sequential()

    
    if d == 1 and a != 'LeakyReLU':
        
        model.add(Dense(b, input_dim=12, activation=a)) 
        model.add(Dense(6, activation='softmax'))

        opt = keras.optimizers.RMSprop(lr=0.5)

        model.compile(optimizer=opt,
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])

        # fit model
        history = model.fit(X_train, y_train, epochs=100, batch_size=c)

        #validate
        accuracy = (history.history['acc'])
        test = np.array(accuracy)
        row = test
    
    elif d == 1 and a == 'LeakyReLU':
        
        model.add(Dense(b, input_dim=12))
        model.add(LeakyReLU(alpha=.001))
        model.add(Dense(6, activation='softmax'))

        opt = keras.optimizers.RMSprop(lr=0.5)

        model.compile(optimizer=opt,
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])

        # fit model
        history = model.fit(X_train, y_train, epochs=100, batch_size=c)

        #validate
        accuracy = (history.history['acc'])
        test = np.array(accuracy)
        row = test
                        
    elif d == 2 and a != 'LeakyReLU':
        
        model.add(Dense(b, input_dim=12, activation=a)) 
        model.add(Dense(b, activation=a)) 
        model.add(Dense(6, activation='softmax'))

        opt = keras.optimizers.RMSprop(lr=0.5)

        model.compile(optimizer=opt,
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])

        # fit model
        history = model.fit(X_train, y_train, epochs=100, batch_size=c)

        #validate
        accuracy = (history.history['acc'])
        test = np.array(accuracy)
        row = test
                        
    elif d == 2 and a == 'LeakyReLU':
        
        model.add(Dense(b, input_dim=12))
        model.add(LeakyReLU(alpha=.001))
        model.add(Dense(b))
        model.add(Dense(6, activation='softmax'))

        opt = keras.optimizers.RMSprop(lr=0.5)

        model.compile(optimizer=opt,
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])

        # fit model
        history = model.fit(X_train, y_train, epochs=100, batch_size=c)

        #validate
        accuracy = (history.history['acc'])
        test = np.array(accuracy)
        row = test
                  
    else:
                  pass
                  
    
    return row
```

```python
#iterate through all combinations of phase 1 hyperparameters
rows = []
for a,b,c,d,in permutations:
    row = create_model_phase_one(X_train, X_test, y_train, y_test,a,b,c,d)
    rows.append(row)
```

```python
#create phase one dataframe
rowtemp = rows
rowtemp = pd.DataFrame(rowtemp) 
```

```python
#construst Phase 1 dataframe 

#rename columns and save as new df
columns = []
for i in range(1,101):
    col = "Epoch "+str(i)
    columns.append(col)
    
rowtemp.columns = columns
Phase1 = rowtemp

#premutations to dataframe
phase_1_permutations = pd.DataFrame(permutations)
phase_1_permutations.columns = ["Activation Function", "Nodes", "Batch Size", "Layers"]

#synthetics data inputs to df 
Learning_Rate = [.05] * len(Phase1)
N_Features = [12] * len(Phase1)
N_Informative = [10] * len(Phase1)
N_Classes = [6] * len(Phase1)
Flip_Y = [.02] * len(Phase1)
Class_Sep = [.05] * len(Phase1)
Shift = [2] * len(Phase1)
Scale = [1] * len(Phase1)
Random_State = [178902] * len(Phase1)
Optimizers = ["Default"] * len(Phase1)
Weight_Initializers = ["Default"] * len(Phase1)
Dropout = [0] * len(Phase1)

Phase_1_Inputs = pd.DataFrame(
    {'Learning_Rate': Learning_Rate,
     'N_Features': N_Features,
     'N_Informative': N_Informative,
     'N_Classes': N_Classes,
     'Flip_Y': Flip_Y,
     'Class_Sep': Class_Sep,
     'Shift': Shift,
     'Scale': Scale,
     'Random_State': Random_State,
     'Optimizers': Optimizers,
     'Weight_Initializers': Weight_Initializers,
     'Dropout': Dropout
     })

#concatenate dfs
Phase_1_DF = pd.concat([Phase_1_Inputs.reset_index(drop=True),
                        phase_1_permutations.reset_index(drop=True),
                        Phase1.reset_index(drop=True)
                       ], axis=1)
```

## Phase 2

```python
# Permutation function 2
def hyper_parameters2(opt, weight, dropout):   
    lists = [opt, weight, dropout]
    permutations2 = list(itertools.product(*lists))
    return permutations2

#lists
Optimizer = ['SGD', 'RMSProp', 'ADAM']
Weight_Initialization = ['glorot_uniform', 'glorot_normal', 'he_uniform', 'he_normal']
Dropout = [0.2, 0.4, 0.5]

#execution of function
permutations2 = hyper_parameters2(Optimizer,Weight_Initialization,Dropout)

#length of all combinations
permutations2
print(len(permutations2))
```

```python
def create_model_phase_two(X_train, X_test, y_train, y_test,a,b,c):
    
    #contruct model
    
    model = models.Sequential()

    
    if a == 'SGD':
        
        model.add(Dense(50, input_dim=12, activation='relu', kernel_initializer=b)) 
        model.add(Dropout(c))
        model.add(Dense(6, activation='softmax'))

        opt = keras.optimizers.SGD(lr=0.5)

        model.compile(optimizer=opt,
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])

        # fit model
        history = model.fit(X_train, y_train, epochs=100, batch_size=32)

        #validate
        accuracy = (history.history['acc'])
        test = np.array(accuracy)
        row = test
    
    elif a == 'RMSProp':
        
        model.add(Dense(50, input_dim=12, activation='relu', kernel_initializer=b)) 
        model.add(Dropout(c))
        model.add(Dense(6, activation='softmax'))

        opt = keras.optimizers.RMSprop(lr=0.5)

        model.compile(optimizer=opt,
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])

        # fit model
        history = model.fit(X_train, y_train, epochs=100, batch_size=32)

        #validate
        accuracy = (history.history['acc'])
        test = np.array(accuracy)
        row = test
                        
    elif a == 'ADAM':
        
        model.add(Dense(50, input_dim=12, activation='relu', kernel_initializer=b)) 
        model.add(Dropout(c))
        model.add(Dense(6, activation='softmax'))

        opt = keras.optimizers.Adam(lr=0.5)

        model.compile(optimizer=opt,
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])

        # fit model
        history = model.fit(X_train, y_train, epochs=100, batch_size=32)

        #validate
        accuracy = (history.history['acc'])
        test = np.array(accuracy)
        row = test
                  
    else:
                  pass
                  
    
    return row
```

```python
#iterate through all combinations of phase 1 hyperparameters
rows2 = []
for a,b,c in permutations2:
    row2 = create_model_phase_two(X_train, X_test, y_train, y_test,a,b,c)
    rows2.append(row2)
    print("permutation done!")
```

```python
phase2temp = pd.DataFrame(rows2)
phase2temp.to_csv('phase2raw.csv')
```

```python
#construst Phase 2 dataframe 

#rename columns and save as new df
columns = []
for i in range(1,101):
    col = "Epoch "+str(i)
    columns.append(col)
    
phase2temp.columns = columns
Phase2 = phase2temp

#premutations to dataframe
phase_2_permutations = pd.DataFrame(permutations2)
phase_2_permutations.columns = ["Optimizers", "Weight_Initializers", "Dropout"]

#synthetics data inputs to df 
Learning_Rate = [.05] * len(Phase2)
N_Features = [12] * len(Phase2)
N_Informative = [10] * len(Phase2)
N_Classes = [6] * len(Phase2)
Flip_Y = [.02] * len(Phase2)
Class_Sep = [.05] * len(Phase2)
Shift = [2] * len(Phase2)
Scale = [1] * len(Phase2)
Random_State = [178902] * len(Phase2)
Batch_Size = [32] * len(Phase2)
Activation_Function = ['relu'] * len(Phase2)
Nodes = [50] * len(Phase2)
Layers = [1] * len(Phase2)

Phase_2_Inputs = pd.DataFrame(
    {'Learning_Rate': Learning_Rate,
     'N_Features': N_Features,
     'N_Informative': N_Informative,
     'N_Classes': N_Classes,
     'Flip_Y': Flip_Y,
     'Class_Sep': Class_Sep,
     'Shift': Shift,
     'Scale': Scale,
     'Random_State': Random_State,
     'Batch Size': Batch_Size,
     'Activation Function': Activation_Function,
     'Nodes': Nodes,
     'Layers': Layers
     })


#concatenate dfs
Phase_2_DF = pd.concat([Phase_2_Inputs.reset_index(drop=True),
                        phase_2_permutations.reset_index(drop=True),
                        Phase2.reset_index(drop=True)
                       ], axis=1)
```

```python
#concat phase 1 and 2
phases = [Phase_1_DF, Phase_2_DF]
FinalDF = pd.concat(phases,sort=True)
```

```python
#reorder columns
col1 = list(Phase_1_DF.columns)
FinalDF = FinalDF[col1]

#export final df
FinalDF.to_csv('JohnCagnoFinalDF.csv')
```
